### Step 1. Go to [Dataset_link](https://raw.githubusercontent.com/APSSDC-Data-Analysis/Data-Analysis-B3-Afternoon/master/D6_Preprocessing(29Aug2020)/titanic.csv)

* This dataset for Titanic passenger details
* Having columns are survived,pclass,name,sex,age,sibsp,parch,ticket,fare,cabin,embarked
* Some passengers were not found in their details (missing values are present)

## Step 2. Use the Csv file and assign it to a dataframe called titan_data

```python
import pandas as pd
titan=pd.read_csv("titanic.csv")
titan_data=pd.DataFrame(titan)
titan_data.head()
```

## Step 3. What is  the number of observations and entries in the dataset? 

```python
print(titan_data.columns)
print(titan_data.index)
```

## Step 4. Check the missing values and non-missing values  for every individual label

```python
for i in list(titan_data.columns):
    print(i,"missing values are",titan_data[i].isnull().sum())
    print(i,"non_missing values are",titan_data[i].notnull().sum())
```

## Step 5. Replace missing values with mean value,If label having  numerics data and assign it to a new dataframe named as  titan_df

```python
import numpy as np
col=[]
for i in list(titan_data.columns):
    if np.dtype(titan_data[i])==np.dtype("int64") or np.dtype(titan_data[i])==np.dtype("float64"):
        col.append(i)
titan_df=pd.DataFrame(titan_data[col].fillna(titan_data[col].mean()))
titan_df
```

## Step 6. In titan_df, Replace missing values with mode value, if label having categorical data (object type) 

```python
col=[]
for i in list(titan_data.columns):
    if np.dtype(titan_data[i])!=np.dtype("int64") and np.dtype(titan_data[i])!=np.dtype("float64"):
        col.append(i)
df=titan_data[col].mode().head()
df
```

```python
titan_df=titan_data[col].replace(to_replace=np.nan,value=df.iloc[0])
titan_df
```

## Step 7. Now, check the missing values in titan_df

```python
titan_df.isnull().sum()
```

## Step 8. Finally, Export titan_df to "cleaned_titan_df.xlsx"

```python
titan_df.to_excel(r'cleaned_titan_df.xlsx', index = True)
```

```python
titan_data=pd.read_csv("titanic.csv")
titan_data.head()
```

```python
from sklearn.impute import SimpleImputer
```

```python
si=SimpleImputer(strategy="mean")
```

```python
col=[]
for i in list(titan_data.columns):
    if np.dtype(titan_data[i])==np.dtype("int64") or np.dtype(titan_data[i])==np.dtype("float64"):
        col.append(i)
titan_df=si.fit_transform(titan_data[col])
titan_df
```

```python
si=SimpleImputer(strategy="constant",fill_value=-1)
titan_df=si.fit_transform(titan_data[col])
titan_df
```

```python
si=SimpleImputer(strategy="most_frequent")
titan_df=si.fit_transform(titan_data)
titan_df
```

```python
titan_df=pd.DataFrame(titan_df,columns=titan_data.columns)
titan_df
```
