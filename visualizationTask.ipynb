## 1. Design a  line plot and scatter plot between 20 random values and its square root in a single plane

```python
import numpy as np
import matplotlib.pyplot as plt
x=[np.random.random(20)]
y=np.sqrt(x)
plt.scatter(x,y)
plt.show()
```

### 2. Draw a scatter plot of 20 equally separated values from 0 - 150 and and its cosine values  with different colors 

```python
x=np.linspace(0,150,20)
y=np.cos(x)
plt.scatter(x,y)
plt.show()
```

### 3. Draw a bar graph between "type" label (in birds.csv file) invertical axis  and its frequency count in horizontal axis and give a proper labels and title

```python
import pandas as pd
birds=pd.read_csv("birds.csv")
birds.head()
```

```python
x=list(birds["type"].unique())
y=list(birds["type"].value_counts())
plt.xlabel("frequency count")
plt.ylabel("type")
plt.title("count of types of birds")
plt.barh(x,y,color=["b","m"])
plt.show()
```

### 5. Take any dataset having outliers  from kaggle and apply preprocessing technique and visualize in box plot

```python
from sklearn.preprocessing import RobustScaler
data=pd.read_csv("datahackers-survey-2019-anonymous-responses.csv")
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# before preprocessing
col=[]
for i in list(data.columns):
    if np.dtype(data[i])==np.dtype("int64") or np.dtype(data[i])==np.dtype("float64"):
        col.append(i)
for i in col:
    data[i].plot(kind="box")
```

```python
rs_data=RobustScaler()
rs_data=rs_data.fit_transform(data[col])
```

```python
rs_data=pd.DataFrame(rs_data,columns=col)
rs_data.head()
```

```python
for i in col:
    rs_data[i].plot(kind="box")
```

```python
r_data=RobustScaler()
r_data=r_data.fit_transform(rs_data[col])
```

```python
r_data=pd.DataFrame(r_data,columns=col)
r_data.head()
```

```python
for i in col:
    r_data[i].plot(kind="box")
```
